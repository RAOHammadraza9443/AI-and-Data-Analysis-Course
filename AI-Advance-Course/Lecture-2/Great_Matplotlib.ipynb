# Line Plot
A line graph is the simplest plot that displays the realtionship between the one independent and one dependent dataset. In this plot , the points are joined by straight line segments.

```python
import numpy as np
X = np.linspace(1,20,100)
Y = np.exp(X)

# Line plot
plt.plot(X, Y)

# Display the plot
plt.show()
```

From the plot, it can be observed that as 'X' is increasing there is an exponential increase in 'Y'.

```python
X = np.linspace(1,20,100)
Y = np.exp(X)

# Line plot
# The argument 'r*' plots each pint as a red "*"
plt.plot(X, Y, "r*")

# Display the plot
plt.show()
```

We can change the colors or shapes of the data points.

There can be multiple line plots in one plot. Let's plot three plots together in a single graph. Also , add a plot title.

```python
X = np.linspace(1,20,100)
Y1 = X
Y2 = np.square(X)
Y3 = np.sqrt(X)

# Line plot
plt.plot(X, Y1 , 'r' , X, Y2, 'b', X, Y3, 'g')

# Add title to the plot
plt.title('Line Plot')

# Display the plot
plt.show()
```

# 1.2 Scatter Plot
A scatter plot is a set of points plotted on horizontal and vertical axes. The scatter plto can be used to study the correlation between the two variables. One can also detect the extreme data points using a scatter plot.

```python
# Check the head() of the laptops dataset
data.head()
```

Plot the scatter plot for the variables "Price" and Rating

```python
# Data
X = data['Price']
Y = data['Rating']

# Plot the scatter plot
plt.scatter(X, Y)

# add the axes labels to plot
plt.xlabel('Price')
plt.ylabel('Rating')

# display the plot
plt.show()
```

We can add different colors , opacity, and shape of data points. Let's add these customizations in the above plot.


```python
# Plot the scatter plot for the variable 'Price' and 'Rating'

X = data['Price']
Y = data['Rating']

# Plot the scatter plot
# s is for shape , C is for colour , alpha is for opacity (0 < alpha < 1)
plt.scatter(X, Y , s = np.array(Y)*2, c = 'green', alpha = 0.4)

# add title
plt.title('Scatter Plot')

# add the axes labels to plot
plt.xlabel('Price')
plt.ylabel('Rating')

# display the plot
plt.show()
```

The bubbles with greater radius display that the data amount is more as compared to the bubbles with less radius.

# 1.6 Box Plot
Boxplot is a way to visualiza the five-number summery of the variable.This five-number summary includes the numerical quantities like minimum, first quartile (Q1),median(Q2),third quartile(Q3), and maximum. Boxplot gives information about the outliers in the data.Detecting and removing outliers is one of the most important steps in exploratory data analysis. Boxplots also tells about the distribution of the data.

```python
data.head()
```

```python
# Plot a distribution of total bill
plt.boxplot(data['Price'])

# Add labels for five number summary
plt.text(x=1.1, y= data['Price'].min(), s='min')
plt.text(x=1.1, y= data.Price.quantile(0.25), s='Q1')
plt.text(x=1.1, y= data['Price'].median(), s='meadian (Q2)')
plt.text(x=1.1, y= data.Price.quantile(0.75), s='Q3')
plt.text(x=1.1, y= data['Price'].max(), s='max')


# Add the graph title and axes labels
plt.title('Boxplot of Total Bill Amount')
plt.ylabel('Price')

# Display the plot
plt.show()
```

```python
# Plot a distribution of total bill
plt.boxplot(data['Price'])

# Add labels for five number summary
plt.text(x=1.1, y= data['Price'].min(), s='min')
plt.text(x=1.1, y= data.Price.quantile(0.25), s='Q1')
plt.text(x=1.1, y= data['Price'].median(), s='meadian (Q2)')
plt.text(x=1.1, y= data.Price.quantile(0.75), s='Q3')
plt.text(x=1.1, y= data['Price'].max(), s='max')

# Add an arrow (annonate) to show the outliers
plt.annotate('Outliers', xy= (0.97, 300000),xytext=(0.7, 300000),
             arrowprops= dict(facecolor = 'red', arrowstyle = 'simple'))
# Add the graph title and axes labels
plt.title('Boxplot of Total Bill Amount')
plt.ylabel('Price')

# Display the plot
plt.show()
```

# 2.5 Distribution Plot

A seaborn provides a distplot() function which is used to visualize a distributon of the univariate variable. This function uses matplotlib to plot a histogram and fit kernel density estimate(KDE).

```python
data
```

Lets plot a distribution plot


```python
# plot a distribution plot
sns.distplot(data['Price'])

plt.show()
```

```python
# Iterate the distplot() function over the primary_storage_type

# list of primary_storage_type
primary_storage_type = ['SSD', 'HDD']

# iterate through time
for i in primary_storage_type:
    subset = data[data['primary_storage_type'] == i]
    
# Draw the density plot
# 'hist = False' will not plot a histogram
# 'kde = True' plots density curve
    sns.distplot(subset['Price'], hist = False, kde = True, 
                kde_kws = {'shade':True},
                label = i)
```

# 2.7 Heatmap

Heatmap is a two-dimensional graphical representation of data where the individual values that are condtained in a matrix are represented as colors. Each square in the heatmap shows the correlation between variable on each axis.
Correlation is a statistic that measures the degree to which two variables move with each other.

```python
data
```

Compute correlation between the variables using .corr() function. Plot a heatmap of the correlation matrix.

```python
corr_matrix = data.corr()

corr_matrix
```

```python
# plot heatmap
# 'annot = True' returns the correlation values
sns.heatmap(corr_matrix, annot = True)

# display the plot
plt.show()
```
