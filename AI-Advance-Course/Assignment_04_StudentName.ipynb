Identify and print pairs of variables with the highest and lowest correlation coefficients

```python
correlation_matrix = df.corr()
correlation_series = correlation_matrix.unstack().sort_values()


correlation_series = correlation_series[correlation_series.index.get_level_values(0) != correlation_series.index.get_level_values(1)]

print("Pairs with the highest correlation coefficients:")
print(correlation_series.tail(1))

print("\nPairs with the lowest correlation coefficients:")
print(correlation_series.head(1))
```

Question 3: Perform the following task: (20 Marks)
Conduct an exploratory data analysis on the Melbourne Housing Market dataset. Identify trends in
housing prices over time, analyze the impact of location, and explore the relationship between
property attributes (such as the number of rooms, type of property, and land size) and price.

```python
melbourne_data = pd.read_csv("/content/Melbourne_housing_FULL.csv")
melbourne_data
```

```python
melbourne_data.info()
```

```python
# Data Preprocessing
# Drop rows with missing values
melbourne_data.dropna(inplace=True)
melbourne_data
```

```python
# Trends in Housing Prices Over Time
# Convert 'Date' column to datetime format
melbourne_data['Date'] = pd.to_datetime(melbourne_data['Date'])
melbourne_data
```

```python
# Extract year and month from the 'Date' column
melbourne_data['Year'] = melbourne_data['Date'].dt.year
melbourne_data['Year']
```

```python
melbourne_data['Month'] = melbourne_data['Date'].dt.month
melbourne_data['Month']
```

```python
# Plot average housing prices over time
average_prices = melbourne_data.groupby(['Year', 'Month'])['Price'].mean()
average_prices.plot(kind='line', figsize=(10, 6))
plt.title('Average Housing Prices Over Time')
plt.xlabel('Year-Month')
plt.ylabel('Average Price')
plt.grid(True)
plt.show()
```

```python
# Impact of Location
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Longtitude', y='Lattitude', hue='Price', data=melbourne_data)
plt.title('Distribution of Housing Prices Across Locations')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

# Relationship Between Property Attributes and Price

```python
# Number of Rooms vs. Price
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Rooms', y='Price', data=melbourne_data)
plt.title('Number of Rooms vs. Price')
plt.xlabel('Number of Rooms')
plt.ylabel('Price')
plt.show()
```

```python
# Type of Property vs. Price
plt.figure(figsize=(10, 6))
sns.boxplot(x='Type', y='Price', data=melbourne_data)
plt.title('Type of Property vs. Price')
plt.xlabel('Type of Property')
plt.ylabel('Price')
plt.show()
```

```python
# Land Size vs. Price
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Landsize', y='Price', data=melbourne_data)
plt.title('Land Size vs. Price')
plt.xlabel('Land Size')
plt.ylabel('Price')
plt.show()
```

```python
# Create a new column indicating passenger age group
def age_group(age):
    if age <= 18:
        return 'Child'
    elif age <= 65:
        return 'Adult'
    else:
        return 'Elderly'

df['Age_Group'] = df['Age'].apply(age_group)
```
