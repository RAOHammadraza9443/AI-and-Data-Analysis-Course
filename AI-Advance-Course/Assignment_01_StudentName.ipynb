1) Create a program that takes a student's numerical grade as input and prints a  corresponding letter grade. Use if-else statements to classify the grade into  categories such as "A," "B," "C," "D," or "F." (Marks 20)
o If the numerical grade is 90 or higher, the corresponding letter grade is "A." o If the numerical grade is between 80 and 89 (inclusive), the corresponding  letter grade is "B."
o If the numerical grade is between 70 and 79 (inclusive), the corresponding  letter grade is "C."
o If the numerical grade is between 60 and 69 (inclusive), the corresponding  letter grade is "D."
o If the numerical grade is below 60, the corresponding letter grade is "F."

```python
student_grade = int(input("Enter the student's numerical grade:"))
if student_grade >= 90:
   letter_grade = "A"
elif student_grade >= 80:
   letter_grade = "B"
elif student_grade >= 70:
   letter_grade = "C"
elif student_grade >= 60:
   letter_grade = "D"
else:
   letter_grade = "F"
print(f"The Corresponding letter grade is: {letter_grade}")
```

2) Create a program that takes a temperature in Celsius as input from user and  classifies it into categories like "Freezing," "Cold," "Moderate," "Warm," or "Hot."  Use if-else statements to define the temperature ranges for each category. (Marks  20)
o If the temperature is below -10 degrees Celsius, it's classified as "Freezing." o If the temperature is between -10 and 0 degrees Celsius (exclusive), it's  classified as "Cold."
o If the temperature is between 0 and 20 degrees Celsius (exclusive), it's  classified as "Moderate."
o If the temperature is between 20 and 30 degrees Celsius (exclusive), it's  classified as "Warm."
o If the temperature is 30 degrees Celsius or above, it's classified as "Hot."

```python
temperature_celsius = int(input("Enter the Temperature Celsius:"))
if temperature_celsius < -10:
  print("Freezing")
elif -10 <= temperature_celsius < 0:
  print("Cold")
elif 0 <=temperature_celsius < 20:
  print("Moderate")
elif 20 <= temperature_celsius < 30:
  print("Warm")
else:
  print("Hot")
```
