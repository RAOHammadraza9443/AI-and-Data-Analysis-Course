Plotting in Python

```python
import pandas as pd
import numpy as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import dataset 
kashti = sns.load_dataset('titanic')
kashti
```

```python
kashti.info()
```

```python
kashti.isnull().sum()
```

```python
kashti.survived.value_counts().plot(kind="bar")
```

```python
sns.countplot(x="survived",hue="sex",data=kashti)
```

```python
# scatter plot
sns.scatterplot(data=kashti,x="age",y="fare",hue="pclass")
```

```python
sns.barplot(data=kashti, x="sex", y="fare",  hue="pclass")
plt.show()
```

```python
kashti.columns
```

```python
sns.pairplot(kashti)
```

```python
kashti.hist(bins=10, figsize=(9,7),grid=False)
```

```python
sns.boxplot(data=kashti, x="sex",y="fare", hue="pclass")
```

```python
sns.boxplot(data=kashti, x="sex",y="fare", hue="who")
```

```python
sns.boxenplot(data=kashti, x="sex",y="fare", hue="pclass")
```

Grid plote

```python
g = sns.FacetGrid(kashti, hue="survived", col="pclass", margin_titles=True,
                  palette={1:"seagreen", 0:"gray"})
g=g.map(plt.scatter, "fare" , "age",edgecolor="w").add_legend();
```

```python
g = sns.FacetGrid(kashti, hue="survived", col="sex", margin_titles=True,
                  palette={1:"seagreen", 0:"gray"})
g=g.map(plt.scatter, "fare" , "age",edgecolor="w").add_legend();
```

```python
sns.stripplot(data=kashti, x="sex",y="fare", hue="class")
```

```python
sns.violinplot(data=kashti, x="sex", y="fare" , hue="class")
```
