```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "tips" dataset
data = sns.load_dataset("tips")

# Create a bar plot of total_bill vs day
sns.barplot(x="day", y="total_bill", data=data)

# Add labels and a title
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill Amount")
plt.title("Average Total Bill Amount by Day")

# Display the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create a sample dataset
data = {
    'Continent': ['Africa', 'Asia', 'Europe', 'North America', 'South America'],
    'Population (billions)': [1.31, 4.61, 0.74, 0.59, 0.43]
}

# Create a DataFrame from the data
df = pd.DataFrame(data)

# Create a bar plot of population by continent
sns.barplot(x='Continent', y='Population (billions)', data=df)

# Add labels and a title
plt.xlabel('Continent')
plt.ylabel('Population (billions)')
plt.title('Population by Continent')

# Rotate the x-axis labels for better readability (optional)
plt.xticks(rotation=45)

# Display the plot
plt.show()
```
